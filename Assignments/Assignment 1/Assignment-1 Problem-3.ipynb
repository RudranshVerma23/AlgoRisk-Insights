```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
ticker = "INFY"
data = yf.download(ticker, start="2023-04-01", end="2024-03-31")

#20 Day SMA using rolling window
data["SMA_20"] = data["Close"].rolling(20).mean()

#10 Day SMA using below code
AVG = data.Close[:10].mean()
data["SMA_10"] = float('NaN')
data.iat[9,7] = AVG
for i in range(10,249):
    data.iat[i,7] = AVG + (data.iat[i,3] - data.iat[i-10,3])/10
    avg = data.iat[i,7]

data
```

```python
plt.plot(data["Close"], label="Closing Price", color="red")
plt.plot(data["SMA_10"], label="10 Day SMA", color="yellow")
plt.plot(data["SMA_20"], label="20 Day SMA", color="cyan")
plt.legend()
```
